```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[5,2,6,8,1,3,4,1,7]
plt.scatter(x,y,marker='*',color='g',s=110)
plt.show()
```

Creating subplot

```python
x=[10,20,30,40,50,60,70,80,90]
y=[3,1,2,4,5,6,7,3,9]
z=[9,6,0,1,5,7,3,7,8]

plt.subplot(1,2,1)
plt.scatter(x,y,marker='*',color='green',s=100)

plt.subplot(1,2,2)
plt.scatter(x,z,marker='*',color='red',s=100)
plt.show()
```
