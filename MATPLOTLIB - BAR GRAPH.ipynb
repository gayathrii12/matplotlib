```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
```

```python
company =['GOOGLE','AMAZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
xposition = np.arange(len(company))      #since we have string in company we are doing this way
xposition
```

```python
plt.xticks(xposition,company) 
#since we get num is xaxis(0,1,2,3) inorder to replcae it with company name we use xticks
plt.ylabel("revenue")
plt.title("US Tech Stocks")
plt.bar(xposition-0.2,revenue,width=0.4,label="revenue")
plt.bar(xposition+0.2,profit,width=0.4,label="Profit")
plt.legend()
```

```python
plt.yticks(xposition,company) 
#since we get num is xaxis(0,1,2,3) inorder to replcae it with company name we use xticks
plt.ylabel("revenue")
plt.title("US Tech Stocks")
plt.barh(xposition-0.2,revenue,label="revenue")
plt.barh(xposition+0.2,profit,label="Profit")
plt.legend()
```
